```python
import yfinance as yf
import pandas as pd

nifty = yf.download('^NSEI', start='2023-01-01', end='2024-12-31')
nifty['daily_return'] = nifty['Close'].pct_change()
nifty['rolling_vol'] = nifty['daily_return'].rolling(30).std()
print(nifty[['Close', 'daily_return', 'rolling_vol']].tail(10))
```
